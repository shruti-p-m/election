```python
pop[['County', 'State']] = pop['Area'].str.split(',', expand=True)
```

```python
pop = pop.drop(columns=['Area', '2020_estimate', 2020, 2021, 2022, 2023])
```

```python
pop_df= pop.iloc[1:]
```

```python
pop_df.head()
```

```python
pop_df.loc[:, 'County'] = (
    pop_df['County']
    .str.replace('.', '', n=1, regex=False)  # Remove only the first period
    .str.replace(' County', '', regex=False)  # Remove the exact phrase ' County'
    .str.replace(' Parish', '', regex=False)
    .str.upper()
    .str.strip()
)

pop_df.loc[:, 'State'] = (
    pop_df['State']
    .str.upper()
    .str.strip()
)
```

```python
pop_df[2024] = pop_df[2024].astype(int)
```

```python
print(pop_df.iloc[:, 0].unique())
```

```python
pop_df.head()
```

```python
pop_df = pop_df.rename(columns={2024: 'Population'})
pop_df.head()
```

```python
df = pd.read_csv('data/cleaned_data.csv')
df['county_fips'] = df['county_fips'].astype(str).str.zfill(5)
df.head()
```

```python
df_2024 = df[df['year'] == 2020].copy().reset_index(drop=True)
df_2024_original = df_2024.copy()
df_2024 = df_2024.drop(columns=['prev_party', 'candidatevotes', 'totalvotes', 'flipped', 'Population'])
df_2024.rename(columns={'party': 'prev_party'}, inplace=True)
df_2024['year'] = 2024
df_2024.head()
```

```python
df_2024_original.head()
```

```python
pop_df = pop_df.rename(columns={
    'County': 'county_name',
    'State': 'state'
})
pop_df.head()
```

```python
merged_df = df_2024.merge(pop_df[['Population', 'county_name', 'state']], 
                      on=['state', 'county_name'], 
                      how='left')
```

```python
merged_df.head()
```

```python
merged_df.isna().sum()
```

```python
merged_df[(merged_df['Population'].isna())].tail(50)
```

```python
# First, set up the key columns for matching
merged_df['merge_key'] = merged_df['county_name'].str.upper().str.strip() + "_" + merged_df['state'].str.upper().str.strip()
df_2024_original['merge_key'] = df_2024_original['county_name'].str.upper().str.strip() + "_" + df_2024_original['state'].str.upper().str.strip()

# Now fill in NaN Population values in merged_df using df_2024_original
merged_df['Population'] = merged_df.apply(
    lambda row: df_2024_original.loc[df_2024_original['merge_key'] == row['merge_key'], 'Population'].values[0]
    if pd.isna(row['Population']) and row['merge_key'] in df_2024_original['merge_key'].values
    else row['Population'],
    axis=1
)

# Optionally, drop the temporary key column
merged_df.drop(columns='merge_key', inplace=True)
df_2024_original.drop(columns='merge_key', inplace=True)
```
